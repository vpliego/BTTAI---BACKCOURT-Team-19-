```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

  df = pd.read_csv('FIFA.CSV')
  print(df.describe())
  print(df.info())
```

```python
df = df.dropna(subset=['Tweet'])
```

```python

import nltk
nltk.download('vader_lexicon')
from nltk.sentiment.vader import SentimentIntensityAnalyzer

analyzer = SentimentIntensityAnalyzer()

def get_sentiment(text):
  scores = analyzer.polarity_scores(text)
  compound_score = scores['compound']
  if compound_score >= 0.05:
    return 1  # Positive
  elif compound_score <= -0.05:
    return -1  # Negative
  else:
    return 0  # Neutral


df['sentiment'] = df['Tweet'].apply(get_sentiment)

df.head(10)
```

```python

average_sentiment = df['sentiment'].mean()
print(f"The average tweet sentiment is: {average_sentiment}")
```

```python
# Filtering the DataFrame to include only rows where the 'Date' starts with '2018-07-02'
filtered_df = df[df['Date'].str.startswith('2018-07-02')]

filtered_df
```

```python

def categorize_tweet(tweet):
  """Categorizes a tweet into one of 5 categories based on the fan experience at the World Cup."""
  tweet = tweet.lower()
  if "atmosphere" in tweet or "stadium" in tweet or "crowd" in tweet or "energy" in tweet:
    return "Stadium Atmosphere"
  elif "hospitality" in tweet or "food" in tweet or "drinks" in tweet or "facilities" in tweet:
    return "Hospitality and Facilities"
  elif "travel" in tweet or "transport" in tweet or "accommodation" in tweet:
    return "Travel and Accommodation"
  elif "security" in tweet or "safety" in tweet or "police" in tweet:
    return "Safety and Security"
  elif "match" in tweet or "game" in tweet or "team" in tweet or "players" in tweet or "score" in tweet:
    return "Match Experience"
  else:
    return "Other"

df['category'] = df['Tweet'].apply(categorize_tweet)

df.head(10)
```

```python

category_counts = df['category'].value_counts()
print(category_counts)
```
