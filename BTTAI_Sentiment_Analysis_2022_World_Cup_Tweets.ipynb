```python

import pandas as pd
df = pd.read_csv('fifa_world_cup_2022_tweets.csv')

```

```python
df.shape
```

```python
df['Numerical Sentiment'] = 0

df.to_csv('updated_fifa_world_cup_2022-tweets.csv', index=False)
```

```python

print(df.columns.tolist())
```

```python

df.loc[df['Sentiment'] == 'positive', 'Numerical Sentiment'] = 1
df.loc[df['Sentiment'] == 'neutral', 'Numerical Sentiment'] = 0
df.loc[df['Sentiment'] == 'negative', 'Numerical Sentiment'] = -1

df.to_csv('updated_fifa_world_cup_2022-tweets.csv', index=False)
```
