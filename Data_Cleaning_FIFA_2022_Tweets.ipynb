```python
# prompt: can you see if any of the dates in the dataset are in 2022

import pandas as pd

df = pd.read_csv('/content/FIFA_World_Cup_Final.csv')

new_df = df[['text', 'created_at', 'lang', 'retweet_count', 'hashtags']]
new_df = new_df[new_df['lang'] == 'en']

# Convert 'created_at' column to datetime objects
new_df['created_at'] = pd.to_datetime(new_df['created_at'])

# Check for dates in 2022
dates_2022 = new_df[new_df['created_at'].dt.year == 2022]

if not dates_2022.empty:
  print("There are dates from 2022 in the dataset.")
  print(dates_2022['created_at'].unique())
else:
  print("There are no dates from 2022 in the dataset.")
```

```python
import nltk
nltk.download('vader_lexicon')
from nltk.sentiment.vader import SentimentIntensityAnalyzer

analyzer = SentimentIntensityAnalyzer()

def get_sentiment(text):
  scores = analyzer.polarity_scores(text)
  compound_score = scores['compound']
  if compound_score >= 0.05:
    return 1  # Positive
  elif compound_score <= -0.05:
    return -1  # Negative
  else:
    return 0  # Neutral


df['sentiment'] = df['Tweet'].apply(get_sentiment)

df.head(10)
```

```python
df_trimmed = df[df['Date'].str.startswith('2018-07-15')]
print(df_trimmed)
```

```python
average_sentiment = df['sentiment'].mean()
print(average_sentiment)
```

```python
df_rts = df_trimmed[df_trimmed['Orig_Tweet'].str.startswith('RT')]

print(f"Original dataset rows: {df_trimmed.shape[0]}")
print(f"Dataset rows with RTs: {df_rts.shape[0]}")
```

```python
df_trimmed_no_rts = df_trimmed[~df_trimmed['Orig_Tweet'].str.startswith('RT')]

print(f"Original dataset rows: {df_trimmed.shape[0]}")
print(f"Dataset rows without RTs: {df_trimmed_no_rts.shape[0]}")
```
